```python
# load the dataset.
movies = pd.read_csv('./dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('./dataset/tmdb_5000_credits.csv')

print(movies.shape, credits.shape)
```

```python
# WE WILL TAKE THE FOLLOWING FROM movies.
    # genres
    # id
    # title
    # original_language
    # keywords
movies = movies[['id', 'title', 'genres', 'original_language', 'overview', 'keywords']]
movies.rename(columns = {'id': 'movie_id'}, inplace = True)
# WE WILL TAKE THE FOLLOWING FROM credits.
    # id
    # cast
    # crew
    # title
credits = credits[['movie_id', 'cast', 'crew']]

# merge them
movies = movies.merge(credits, on='movie_id')
```

```python
# let's unwrap from string values.
import ast


def get_genres_keywords(string):
    genreList = []
    for i in ast.literal_eval(string):
        genreList.append(i.get('name')) 
    return genreList

def get_cast(string):
    # let's show top 2 casting.
    cast_list = []
    count = 0
    for i in ast.literal_eval(string):
        cast_list.append(i.get("name"))
        count += 1
        if count == 2:
            break
    return cast_list

def get_director(string):
    dir_list = []
    for i in ast.literal_eval(string):
        if i.get("job").lower() == "director":
            dir_list.append(i.get("name"))
    return dir_list

def remove_spaces(List):
    new = []
    for i in List:
        new.append(i.replace(' ', ''))
    return new

movies['genres'] = movies['genres'].apply(lambda x: get_genres_keywords(x))
movies['keywords'] = movies['keywords'].apply(lambda x: get_genres_keywords(x))
movies['cast'] = movies['cast'].apply(lambda x: get_cast(x))
movies['crew'] = movies['crew'].apply(lambda x: get_director(x))

movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
```

```python
# unfold the lists
movies['genres'] = movies['genres'].map(lambda x: " ".join(x))
movies['keywords'] = movies['keywords'].map(lambda x: " ".join(x))
movies['cast'] = movies['cast'].map(lambda x: " ".join(x))
movies['crew'] = movies['crew'].map(lambda x: " ".join(x))
```

```python
movies.sample(5)
```

```python
# create the final tag column on which we will build our model
movies['final_tag'] = movies['overview'] + ' ' + movies['original_language'] + ' ' + movies['keywords']
```

```python
final_df = movies[['title', 'cast', 'crew', 'genres', 'final_tag', 'movie_id']]
```

```python
final_df.rename(columns = {'title': 'Title'}, inplace = True)
```
